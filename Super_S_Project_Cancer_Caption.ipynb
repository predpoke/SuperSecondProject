```python
#토큰화

SPECIAL_TOKENS = ["<pad>", "<sos>", "<eos>", "<unk>"]


def tokenize(text):
  return str(text).strip().split()

```

```python
#단어 사전 만들기

vocab = {}

for token in SPECIAL_TOKENS :
  vocab[token] = len(vocab)


all_texts = list(df["src_text"]) + list(df["tgt_text"])

for text in all_texts :
  for word in tokenize(text) :
    if word not in vocab :
      vocab[word] = len(vocab)


word2idx = vocab
idx2word = {v: k for k, v in word2idx.items()}


pad_idx = word2idx["<pad>"]
sos_idx = word2idx["<sos>"]
eos_idx = word2idx["<eos>"]
unk_idx = word2idx["<unk>"]

print("vocab size :", len(vocab))
```

```python
def encode_src(text) :
  tokens = tokenize(text)
  ids = [word2idx.get(tok, unk_idx) for tok in tokens]
  return ids


def encode_tgt(text) :
  tokens = tokenize(text)
  ids = [sos_idx] + [word2idx.get(tok, unk_idx) for tok in tokens] + [eos_idx]
  return ids

```

```python
import torch
from torch.utils.data import Dataset, DataLoader

class SkinCaptionTextDataset(Dataset):
  def __init__(self, dataframe):
    self.df = dataframe.reset_index(drop=True)

  def __len__(self):
    return len(self.df)

  def __getitem__(self, idx) :
    row = self.df.iloc[idx]


    src_text = row["src_text"]
    tgt_text = row["tgt_text"]

    src_ids = encode_src(src_text)
    tgt_ids = encode_tgt(tgt_text)


    return {
        "src_ids" : torch.tensor(src_ids, dtype=torch.long),
        "tgt_ids" : torch.tensor(tgt_ids, dtype=torch.long)
    }

    #DataLoader에 넣을 수 있는 형태로 만든다
```

```python
from torch.nn.utils.rnn import pad_sequence

def collate_fn(batch):
  src_batch = [item["src_ids"] for item in batch]
  tgt_batch = [item["tgt_ids"] for item in batch]

  src_batch = pad_sequence(
      src_batch,
      batch_first=True,
      padding_value=pad_idx
  )


  tgt_batch = pad_sequence(
      tgt_batch,
      batch_first=True,
      padding_value=pad_idx
  )


  return src_batch, tgt_batch

  #padding 함수 만들기
  # 여러 문장 길이를 맞추려고 뒤에 <pad> 붙인느 것
```

```python
train_dataset = SkinCaptionTextDataset(train_df)
val_dataset = SkinCaptionTextDataset(val_df)
test_dataset = SkinCaptionTextDataset(test_df)


train_loader= DataLoader(
    train_dataset,
    batch_size= 8,
    shuffle=True,
    collate_fn = collate_fn
)


val_loader = DataLoader(
    val_dataset,
    batch_size =8,
    shuffle=False,
    collate_fn = collate_fn
)


test_loader = DataLoader(
    test_dataset,
    batch_size=8,
    shuffle=False,
    collate_fn = collate_fn
)
```

```python
#여기서부터 모델 코드


import torch.nn as nn
import math


class PositionalEncoding(nn.Module) :
  def __init__(self, d_model, max_len=5000) :
    super().__init__()


    pe = torch.zeros(max_len, d_model)

    position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)

    div_term = torch.exp(
        torch.arange(0, d_model, 2).float()
        * (-math.log(10000.0) / d_model)

    )

    pe[:, 0::2] = torch.sin(position * div_term)

    pe[:, 1::2] = torch.cos(position * div_term)

    pe = pe.unsqueeze(0)

    self.register_buffer("pe", pe)


  def forward(self, x) :
    seq_len = x.size(1)
    return x +self.pe[:, :seq_len]
```

```python
class TextToTextTransformer(nn.Module) :
  def __init__(
      self,
      vocab_size,
      d_model=128,
      nhead=4,
      num_encoder_layers=2,
      num_decoder_layers=2,
      dim_feedforward=512,
      dropout=0.1,
      pad_idx=0
  ):
      super().__init__()

      self.pad_idx = pad_idx
      self.d_model = d_model


      self.src_embedding = nn.Embedding(vocab_size, d_model, padding_idx=pad_idx)
      self.tgt_embedding = nn.Embedding(vocab_size, d_model, padding_idx=pad_idx)

      self.pos_encoding = PositionalEncoding(d_model)

      self.transformer = nn.Transformer(
          d_model=d_model,
          nhead=nhead,
          num_encoder_layers=num_encoder_layers,
          num_decoder_layers=num_decoder_layers,
          dim_feedforward=dim_feedforward,
          dropout=dropout,
          batch_first=True
          )
      self.fc_out = nn.Linear(d_model, vocab_size)

  def make_tgt_mask(self, tgt) :
    tgt_len = tgt.size(1)

    mask = torch.triu(
        torch.ones(tgt_len, tgt_len),
        diagonal=1
    )

    return mask.to(tgt.device)

  def forward(self, src, tgt_input) :
    src_key_padding_mask = (src == self.pad_idx)
    tgt_key_padding_mask = (tgt_input == self.pad_idx)

    tgt_mask = self.make_tgt_mask(tgt_input)

    src_emb = self.src_embedding(src)* math.sqrt(self.d_model)
    tgt_emb = self.tgt_embedding(tgt_input)* math.sqrt(self.d_model)

    src_emb = self.pos_encoding(src_emb)
    tgt_emb = self.pos_encoding(tgt_emb)

    output = self.transformer(
        src = src_emb,
        tgt=tgt_emb,
        tgt_mask = tgt_mask,
        src_key_padding_mask = src_key_padding_mask,
        tgt_key_padding_mask = tgt_key_padding_mask,
        memory_key_padding_mask = src_key_padding_mask
    )

    logits = self.fc_out(output)

    return logits


```

```python
#디코더에는 앞부분을 넣고 다음 단어를 맞히게 해야한다
#그래서자를때 [:, :-1 이런식으로자르는 것 ]
```

```python
def train_one_epoch(model, train_loader, optimizer, criterion, device) :
  model.train()

  total_loss = 0

  for src, tgt in train_loader :
    src = src.to(device)
    tgt = tgt.to(device)

    tgt_input = tgt[:, :-1]


    tgt_output = tgt[:, 1:]

    logits = model(src, tgt_input)

    loss = criterion(
        logits.reshape(-1, logits.size(-1)),
        tgt_output.reshape(-1)
    )

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    total_loss += loss.item()

  return total_loss / len(train_loader)
```

```python
def evaluate(model, val_loader, criterion, device) :
  model.eval()

  total_loss =0

  with torch.no_grad():
    for src, tgt in val_loader :
      src = src.to(device)
      tgt = tgt.to(device)

      tgt_input = tgt[:, :-1]
      tgt_output = tgt[:, 1:]

      logits = model(src, tgt_input)


      loss = criterion(
          logits.reshape(-1, logits.size(-1)),
          tgt_output.reshape(-1)
      )

      total_loss += loss.item()

  return total_loss/ len(val_loader)
```

```python
epochs =30

for epoch in range(epochs) :
  train_loss = train_one_epoch(
      model,
      train_loader,
      optimizer,
      criterion,
      device
  )

  val_loss = evaluate(
      model,
      val_loader,
      criterion,
      device
)

  print(f"Epoch{epoch+1}/{epochs} | Train Loss : {train_loss:.4f} | Val Loss : {val_loss:.4f}")
```

```python
def count_unk_ratio(dataframe) :
  total = 0
  unk = 0

  for text in list(dataframe["src_text"]) + list(dataframe["tgt_text"]):
    for tok in tokenize(text):
      total += 1
      if tok not in word2idx :
                   unk += 1
  return unk, total, unk /total


train_unk, train_total, train_ratio = count_unk_ratio(train_df)
val_unk, val_total, val_ratio = count_unk_ratio(val_df)

print("train unk:", train_unk, "/" ,train_total, "ratio", train_ratio)
print("val unk", val_unk, "/", val_total, "ratio", val_ratio)
```

```python
def generate_text(model, src_text, max_len=80) :
  model.eval()

  src_idx = encode_src(src_text)
  src = torch.tensor(src_idx, dtype=torch.long).unsqueeze(0).to(device)

  generated = [sos_idx]

  with torch.no_grad():
    for _ in range(max_len) :
      tgt_input = torch.tensor(generated, dtype = torch.long).unsqueeze(0).to(device)

      logits = model(src, tgt_input)

      next_token_logits = logits[0, -1, :]
      next_token_id = next_token_logits.argmax().item()

      generated.append(next_token_id)

      if next_token_id == eos_idx:
        break

  words = []

  for idx in generated :
    if idx in [sos_idx, eos_idx, pad_idx]:
      continue
    words.append(idx2word.get(idx, "<unk>"))

  return " ".join(words)
```

```python
#df_ham에 들어있는 900개의 SkincAP에데이터에 맞는 이미지를 꺼내서 새 폴더에 저장하고 HAM 형식의 metadata를 새로 만드는 작업


import os
import shutil
from tqdm import tqdm
import pandas as pd

output_base = "/content/drive/MyDrive/datasets/SkinCAP_HAM_format"
output_image_dir = os.path.join(output_base, "images")

os.makedirs(output_image_dir, exist_ok=True)

copied_rows = []
missing_files = []

for idx, row in tqdm(df_ham.iterrows(), total=len(df_ham)):
    image_name = row["image_name"]

    src_path = os.path.join(source_image_dir, image_name)
    dst_path = os.path.join(output_image_dir, image_name)

    if os.path.exists(src_path):
        shutil.copy2(src_path, dst_path)

        copied_rows.append({
            "image_id": os.path.splitext(image_name)[0],
            "image_name": image_name,
            "image_path": dst_path,
            "dx": row["ham_label"],
            "original_label": row["label_text"],
            "caption": row["caption_en"],
            "source": "SkinCAP"
        })
    else:
        missing_files.append(image_name)

print("복사 성공:", len(copied_rows))
print("누락 이미지:", len(missing_files))
print("누락 예시:", missing_files[:20])
```

```python
skincap_ham_df = pd.DataFrame(copied_rows)

print(skincap_ham_df.head())
print(skincap_ham_df["dx"].value_counts())

metadata_path = os.path.join(output_base, "skincap_ham_metadata.csv")

skincap_ham_df.to_csv(metadata_path, index=False, encoding="utf-8-sig")

print("저장 완료:", metadata_path)
```

```python

import os
#여기서부터 이미지 가공된거 저장한느 곳인데 일단 앞에거 실행후 해야할 듯

output_base = "/content/drive/MyDrive/datasets/SkinCAP_HAM_format"
output_image_dir = os.path.join(output_base, "images")

os.makedirs(output_image_dir,exist_ok=True)

copied_rows = []
missing_files =[]

for idx, row in tqdm(df_ham.iterrows(), total=len(df_ham)):
  image_name = str(row["image_name"]).strip()

  try :
    image_index = int(os.path.splitext(image_name)[0]) -1

  except :
    missing_files.append(image_name)
    continue

  if image_index < 0 or image_index >= len(image_data) :
    missing_files.append(image_name)
    continue

  img = image_data[image_index]['image']

  dst_path = os.path.join(output_image_dir, image_name)


  img.save(dst_path)

  copied_rows.append({
      "image_id" : os.path.splitext(image_name)[0],
      "image_name" : image_name,
      "image_path" : dst_path,
      "dx" : row["ham_label"],
      "original_label" : row["label_text"],
      "caption" : row["caption_en"],
      "source" : "SkinCAP"
  })

print("저장 성공 :", len(copied_rows))
print("누락 :", len(missing_files))
print("누락 예시 :", missing_files[:20])

```
