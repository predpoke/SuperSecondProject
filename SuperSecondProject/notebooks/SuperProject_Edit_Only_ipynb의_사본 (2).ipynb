```python
image_exts = (".jpg", ".jpeg", ".png", ".bmp", ".webp") #이런 파일만 인정시킴

temp_dataset = datasets.ImageFolder(root=data_dir)
class_to_idx = temp_dataset.class_to_idx

rows = []
bad_files = []

for class_name, label in tqdm(class_to_idx.items(), desc="metadata 생성"): #items 가 이 내부 데이터를 쌍으로 꺼내줌 키 밸류 이런식으로
    class_dir = os.path.join(data_dir, class_name)

    if not os.path.isdir(class_dir):
        continue #실제 폴더인지 확인함

    for filename in os.listdir(class_dir):
        if filename.lower().endswith(image_exts):
            image_path = os.path.join(class_dir, filename)

            try:
                with Image.open(image_path) as img:
                    img.verify() #검사한다

                rows.append({
                    "image_path": image_path,
                    "filename": filename,
                    "class_name": class_name,
                    "label": label
                })

            except Exception as e:
                bad_files.append({
                    "image_path": image_path,
                    "filename": filename,
                    "class_name": class_name,
                    "label": label,
                    "error": str(e)
                })

metadata_df = pd.DataFrame(rows)
bad_df = pd.DataFrame(bad_files)

print("정상 이미지 수:", len(metadata_df))
print("문제 이미지 수:", len(bad_df))
print("클래스 수:", metadata_df["class_name"].nunique())
print(metadata_df.head())

metadata_csv_path = "/content/drive/MyDrive/datasets/officehome_metadata_cnn_clean.csv"
bad_csv_path = "/content/drive/MyDrive/datasets/officehome_bad_images.csv"

metadata_df.to_csv(metadata_csv_path, index=False)
bad_df.to_csv(bad_csv_path, index=False)

print("metadata 저장:", metadata_csv_path)
print("bad images 저장:", bad_csv_path)
```

```python
class OfficeHomeDualViewDataset(Dataset):
    def __init__(self, csv_path, transform_global=None, transform_local=None):
        self.df = pd.read_csv(csv_path)
        self.transform_global = transform_global
        self.transform_local = transform_local

    def __len__(self): #Dataloader가 전체 데이터 개수를 알아야 batch 를 만들 수 있어서 꼭 필요함
        return len(self.df)

    def __getitem__(self, idx): # idx 번째 데이터를 꺼내는 함수
        row = self.df.iloc[idx]

        image_path = row["image_path"]
        label = int(row["label"])

        image = Image.open(image_path).convert("RGB")


        if self.transform_global :
          image_global = self.transform_global(image)
        else :
          image_global = image

        if self.transform_local :
          image_local = self.transform_local(image)
        else :
          image_local = image



        return image_global, image_local, label

class OfficeHomeCSVDataset(Dataset):
    def __init__(self, csv_path, transform=None, return_path=False):
        self.df = pd.read_csv(csv_path)
        self.transform = transform
        self.return_path = return_path

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        image_path = row["image_path"]
        label = int(row["label"])

        image = Image.open(image_path).convert("RGB")

        if self.transform:
            image = self.transform(image)

        if self.return_path:
            return image, label, image_path

        return image, label
```

```python
#6.15 추가사항
import torch.nn.functional as F

def consistency_kl_loss(logits_global, logits_local, temperature=1.0):
    p_global = F.softmax(logits_global / temperature, dim=1).detach()
    log_p_local = F.log_softmax(logits_local / temperature, dim=1) #여기서 temperature 는 softmax 분포를 부드럽게 만들거나 날카롭게 만든느 값
    #dim=1은 class 방향으로 계산 0이면 batch 방향

 #logit : 모델이 softmax를 거치기 전에 뱉는 원점수 그니까 날것 그대로의 점수 여기에 softmax를 적용하면 이제 확률이 됨
    loss = F.kl_div(
        log_p_local,
        p_global,
        reduction="batchmean"
    ) * (temperature ** 2)

    return loss



def train_one_epoch_mixup_global_consistency(
    model,
    train_loader,
    criterion,
    optimizer,
    device,
    alpha=0.2,
    consistency_weight=0.2,
    temperature=1.0
):
    model.train()

    running_loss = 0.0
    running_mixup_loss = 0.0
    running_consistency_loss = 0.0

    correct_ref = 0
    total = 0

    start_time = time.time()

    progress_bar = tqdm(
        train_loader,
        desc=f"Training MixUp + GlobalLocal Consistency alpha={alpha}",
        leave=True
    )

    for batch_idx, (images_global, images_local, labels) in enumerate(progress_bar):
        batch_start = time.time()

        images_global = images_global.to(device, non_blocking=True)
        images_local = images_local.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        # MixUp은 global view 기준으로 적용
        mixed_images, labels_a, labels_b, lam = mixup_data(
            images_global,
            labels,
            alpha=alpha
        )

        optimizer.zero_grad(set_to_none=True)

        with autocast():
            # 1. MixUp classification loss
            outputs_mix = model(mixed_images)

            loss_mix = mixup_criterion(
                criterion,
                outputs_mix,
                labels_a,
                labels_b,
                lam
            )
# 2. Global-Local consistency

            # 같은 이미지의 두 view를 비교함

            with torch.no_grad():
              outputs_global = model(images_global)
            outputs_local = model(images_local)

            loss_consistency = consistency_kl_loss(
                logits_global=outputs_global,
                logits_local=outputs_local,
                temperature=temperature
            )

            loss = loss_mix + consistency_weight * loss_consistency

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        running_loss += loss.item() * labels.size(0)
        running_mixup_loss += loss_mix.item() * labels.size(0)
        running_consistency_loss += loss_consistency.item() * labels.size(0)

        # MixUp acc는 참고값
        _, preds = torch.max(outputs_mix, 1)
        correct_ref += (preds == labels_a).sum().item()
        total += labels.size(0)

        current_loss = running_loss / total
        current_mixup_loss = running_mixup_loss / total
        current_consistency_loss = running_consistency_loss / total
        current_acc_ref = correct_ref / total
        batch_time = time.time() - batch_start

        progress_bar.set_postfix({
            "loss": f"{current_loss:.4f}",
            "mix": f"{current_mixup_loss:.4f}",
            "cons": f"{current_consistency_loss:.4f}",
            "acc_ref": f"{current_acc_ref:.4f}",
            "sec/batch": f"{batch_time:.2f}"
        })

    epoch_time = time.time() - start_time

    epoch_loss = running_loss / total
    epoch_acc_ref = correct_ref / total
    epoch_mixup_loss = running_mixup_loss / total
    epoch_consistency_loss = running_consistency_loss / total

    return (
        epoch_loss,
        epoch_acc_ref,
        epoch_mixup_loss,
        epoch_consistency_loss,
        epoch_time
    )
```

```python
#6.14 추가사항
def mixup_data(images, labels, alpha=0.2):
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1.0

    batch_size = images.size(0)
    index = torch.randperm(batch_size).to(images.device)

    mixed_images = lam * images + (1 - lam) * images[index]

    labels_a = labels
    labels_b = labels[index]

    return mixed_images, labels_a, labels_b, lam


def mixup_criterion(criterion, outputs, labels_a, labels_b, lam):
    loss_a = criterion(outputs, labels_a)
    loss_b = criterion(outputs, labels_b)

    return lam * loss_a + (1 - lam) * loss_b


```

```python
#6.14 추가사항
def evaluate_clean_train_acc(
    model,
    train_loader,
    criterion,
    device,
    max_batches=None
):
    model.eval()

    running_loss = 0.0
    correct = 0
    total = 0

    start_time = time.time()

    with torch.no_grad():
        for batch_idx, (images, labels) in enumerate(tqdm(train_loader, desc="Clean Train Eval", leave=True)):
            if max_batches is not None and batch_idx >= max_batches:
                break

            images = images.to(device)
            labels = labels.to(device)

            with autocast():
                outputs = model(images)
                loss = criterion(outputs, labels)

            running_loss += loss.item() * images.size(0)

            _, preds = torch.max(outputs, 1)

            correct += (preds == labels).sum().item()
            total += labels.size(0)

    epoch_time = time.time() - start_time

    clean_loss = running_loss / total
    clean_acc = correct / total

    return clean_loss, clean_acc, epoch_time



def add_dropout_to_convnext_classifier(model, dropout_p=0.1):
    old_head = model.classifier[2]

    if isinstance(old_head, nn.Sequential):
        print("classifier가 이미 Sequential입니다. Dropout 추가 생략")
        return model

    if not isinstance(old_head, nn.Linear):
        raise TypeError(f"예상한 Linear가 아닙니다: {type(old_head)}")

    in_features = old_head.in_features
    out_features = old_head.out_features

    new_head = nn.Sequential(
        nn.Dropout(p=dropout_p),
        nn.Linear(in_features, out_features)
    )

    with torch.no_grad():
        new_head[1].weight.copy_(old_head.weight)
        new_head[1].bias.copy_(old_head.bias)

    model.classifier[2] = new_head

    print(f"Dropout 추가 완료: p={dropout_p}")

    return model
```

```python
#6.14
def freeze_convnext_front_keep_last(model):
    for param in model.parameters():
        param.requires_grad = False

    for param in model.features[-1].parameters():
        param.requires_grad = True

    for param in model.classifier.parameters():
        param.requires_grad = True

    trainable_count = sum(
        p.numel() for p in model.parameters()
        if p.requires_grad
    )
    total_count = sum(p.numel() for p in model.parameters())

    print("Phase 2 freeze 완료")
    print("학습 가능 파라미터 수:", trainable_count)
    print("전체 파라미터 수:", total_count)
    print("학습 가능 비율:", round(trainable_count / total_count, 4))

    return model
```

```python
#6.18추가사항
class CenterLoss(nn.Module):
    def __init__(self, num_classes, feat_dim, device):
        super().__init__()

        self.num_classes = num_classes
        self.feat_dim = feat_dim

        self.centers = nn.Parameter(
            torch.randn(num_classes, feat_dim).to(device)
        )

    def forward(self, features, labels):
        """
        features: [batch, feat_dim]
        labels  : [batch]
        """
        batch_centers = self.centers[labels]

        loss = ((features - batch_centers) ** 2).sum(dim=1).mean() / 2.0

        return loss


def convnext_forward_with_features(model, images):
    """
    ConvNeXt에서 classifier 직전 feature와 logits를 함께 반환.
    Dropout을 classifier에 추가한 경우도 대응.
    """

    x = model.features(images)
    x = model.avgpool(x)

    # classifier[0] = LayerNorm2d
    # classifier[1] = Flatten
    # classifier[2] = Linear 또는 Sequential(Dropout, Linear)
    x = model.classifier[0](x)
    features = model.classifier[1](x)

    logits = model.classifier[2](features)

    return logits, features


def get_convnext_feature_dim(model):
    head = model.classifier[2]

    if isinstance(head, nn.Linear):
        return head.in_features

    if isinstance(head, nn.Sequential):
        for layer in head:
            if isinstance(layer, nn.Linear):
                return layer.in_features

    raise TypeError(f"Linear head를 찾을 수 없습니다: {type(head)}")



def train_one_epoch_centerloss_dualview(
    model,
    train_loader,
    criterion,
    center_loss_fn,
    optimizer,
    center_optimizer,
    device,
    center_loss_weight=0.02
):
    model.train()
    center_loss_fn.train()

    running_loss = 0.0
    running_ce_loss = 0.0
    running_center_loss = 0.0

    correct = 0
    total = 0

    start_time = time.time()

    progress_bar = tqdm(
        train_loader,
        desc=f"Training CenterLoss Phase weight={center_loss_weight}",
        leave=True
    )

    for batch_idx, (images_global, images_local, labels) in enumerate(progress_bar):
        batch_start = time.time()

        # Phase 2에서는 global view만 사용
        images = images_global.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)
        center_optimizer.zero_grad(set_to_none=True)

        with autocast():
            logits, features = convnext_forward_with_features(
                model,
                images
            )

            ce_loss = criterion(logits, labels)

            c_loss = center_loss_fn(
                features,
                labels
            )

            loss = ce_loss + center_loss_weight * c_loss

        scaler.scale(loss).backward()

        scaler.step(optimizer)
        scaler.step(center_optimizer)
        scaler.update()

        running_loss += loss.item() * labels.size(0)
        running_ce_loss += ce_loss.item() * labels.size(0)
        running_center_loss += c_loss.item() * labels.size(0)

        _, preds = torch.max(logits, 1)

        correct += (preds == labels).sum().item()
        total += labels.size(0)

        current_loss = running_loss / total
        current_ce = running_ce_loss / total
        current_center = running_center_loss / total
        current_acc = correct / total
        batch_time = time.time() - batch_start

        progress_bar.set_postfix({
            "loss": f"{current_loss:.4f}",
            "ce": f"{current_ce:.4f}",
            "center": f"{current_center:.4f}",
            "acc": f"{current_acc:.4f}",
            "sec/batch": f"{batch_time:.2f}"
        })

    epoch_time = time.time() - start_time

    epoch_loss = running_loss / total
    epoch_acc = correct / total
    epoch_ce_loss = running_ce_loss / total
    epoch_center_loss = running_center_loss / total

    return (
        epoch_loss,
        epoch_acc,
        epoch_ce_loss,
        epoch_center_loss,
        epoch_time
    )
```

```python
#6.17추가부분
# 점수 낮은 애들을 위한 필터 만들어서 적용

def get_hard_classes_from_val(
    val_labels,
    val_preds,
    classes,
    acc_threshold=0.82,
    max_classes=10,
    priority_names=None
):
    class_acc_df = make_class_accuracy_table(
        labels=val_labels,
        preds=val_preds,
        classes=classes
    )

    hard_df = class_acc_df[
        class_acc_df["accuracy"] < acc_threshold
    ].copy()

    hard_df = hard_df.sort_values("accuracy", ascending=True)

    hard_class_names = hard_df["class_name"].head(max_classes).tolist()

    # 전자기기 계열은 우선 포함
    if priority_names is not None:
        for name in priority_names:
            if name in classes and name not in hard_class_names:
                hard_class_names.append(name)

    return hard_class_names, class_acc_df


from torch.utils.data import WeightedRandomSampler

def make_hard_class_sampler(
    dataset,
    hard_class_names,
    class_to_idx,
    hard_sample_weight=4.0,
    normal_sample_weight=1.0
):
    hard_label_ids = set([
        class_to_idx[name]
        for name in hard_class_names
        if name in class_to_idx
    ])

    sample_weights = []

    for _, row in dataset.df.iterrows():
        label = int(row["label"])

        if label in hard_label_ids:
            sample_weights.append(hard_sample_weight)
        else:
            sample_weights.append(normal_sample_weight)

    sampler = WeightedRandomSampler(
        weights=sample_weights,
        num_samples=len(sample_weights),
        replacement=True
    )

    return sampler


def make_hard_class_weight_tensor(
    num_classes,
    classes,
    hard_class_names,
    device,
    hard_loss_weight=1.4,
    normal_loss_weight=1.0
):
    weights = torch.ones(num_classes) * normal_loss_weight

    for name in hard_class_names:
        if name in classes:
            idx = classes.index(name)
            weights[idx] = hard_loss_weight

    weights = weights.to(device)

    return weights



def train_one_epoch_mixup_dualview_only(
    model,
    train_loader,
    criterion,
    optimizer,
    device,
    alpha=0.15
):
    model.train()

    running_loss = 0.0
    correct_ref = 0
    total = 0

    start_time = time.time()

    progress_bar = tqdm(
        train_loader,
        desc=f"Training Hard Phase MixUp alpha={alpha}",
        leave=True
    )

    for batch_idx, (images_global, images_local, labels) in enumerate(progress_bar):
        batch_start = time.time()

        images = images_global.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        mixed_images, labels_a, labels_b, lam = mixup_data(
            images,
            labels,
            alpha=alpha
        )

        optimizer.zero_grad(set_to_none=True)

        with autocast():
            outputs = model(mixed_images)

            loss = mixup_criterion(
                criterion,
                outputs,
                labels_a,
                labels_b,
                lam
            )

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        running_loss += loss.item() * labels.size(0)

        _, preds = torch.max(outputs, 1)
        correct_ref += (preds == labels_a).sum().item()
        total += labels.size(0)

        current_loss = running_loss / total
        current_acc_ref = correct_ref / total
        batch_time = time.time() - batch_start

        progress_bar.set_postfix({
            "loss": f"{current_loss:.4f}",
            "acc_ref": f"{current_acc_ref:.4f}",
            "sec/batch": f"{batch_time:.2f}"
        })

    epoch_time = time.time() - start_time

    return running_loss / total, correct_ref / total, epoch_time
```

```python
from torch.cuda.amp import autocast, GradScaler

scaler = GradScaler()

def train_one_epoch_mixup(model, train_loader, criterion, optimizer, device, alpha=0.2):
    model.train()

    running_loss = 0.0
    correct = 0
    total = 0
    top3_sum = 0.0
    batch_count = 0

    start_time = time.time()

    progress_bar = tqdm(train_loader, desc="Training MixUp", leave=True)

    for batch_idx, (images, labels) in enumerate(progress_bar):
        batch_start = time.time()

        images = images.to(device)
        labels = labels.to(device)

        images, labels_a, labels_b, lam = mixup_data(
            images,
            labels,
            alpha = alpha
        )

        optimizer.zero_grad()


        with autocast():
          outputs = model(images)
          loss = mixup_criterion(
              criterion,
              outputs,
              labels_a,
              labels_b,
              lam
              )

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()


        running_loss += loss.item() * images.size(0)

        _, preds = torch.max(outputs, 1)
        correct += (preds == labels_a).sum().item()
        total += labels.size(0)

        current_loss = running_loss / total
        current_acc = correct / total
        batch_time = time.time() - batch_start

        progress_bar.set_postfix({
            "batch": batch_idx + 1,
            "loss": f"{current_loss:.4f}",
            "acc": f"{current_acc:.4f}",
            "sec/batch": f"{batch_time:.2f}"
        })

    epoch_time = time.time() - start_time

    epoch_loss = running_loss / total
    epoch_acc = correct / total

    return epoch_loss, epoch_acc, epoch_time
```

```python
def topk_accuracy(outputs, labels, k=3):
    with torch.no_grad():
        _, topk_preds = outputs.topk(k, dim=1)
        correct = topk_preds.eq(labels.view(-1, 1).expand_as(topk_preds))
        return correct.any(dim=1).float().mean().item()
```

```python
def evaluate(model, val_loader, criterion, device):
    model.eval()

    running_loss = 0.0
    correct = 0
    total = 0

    top3_sum = 0.0
    batch_count = 0

    all_preds = []
    all_labels = []

    start_time = time.time()

    progress_bar = tqdm(val_loader, desc="Validation", leave=True)

    with torch.no_grad():
        for batch_idx, (images, labels) in enumerate(progress_bar):
            batch_start = time.time()

            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            running_loss += loss.item() * images.size(0)

            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

            top3 = topk_accuracy(outputs, labels, k=3)
            top3_sum += top3
            batch_count += 1

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

            current_loss = running_loss / total
            current_acc = correct / total
            current_top3 = top3_sum / batch_count
            batch_time = time.time() - batch_start

            progress_bar.set_postfix({
                "batch": batch_idx + 1,
                "loss": f"{current_loss:.4f}",
                "acc": f"{current_acc:.4f}",
                "top3" : f"{current_top3:.4f}",
                "sec/batch": f"{batch_time:.2f}"
            })

    epoch_time = time.time() - start_time

    epoch_loss = running_loss / total
    epoch_acc = correct / total
    epoch_top3 = top3_sum / batch_count

    return epoch_loss, epoch_acc, epoch_top3, all_preds, all_labels, epoch_time
```

```python
def make_class_accuracy_table(labels, preds, classes) :
  labels = np.array(labels)
  preds = np.array(preds)

  rows = []

  for idx, class_name in enumerate(classes):
    mask = labels == idx
    total = mask.sum()

    if total == 0:
      correct = 0
      acc = 0.0

    else :
      correct = int((preds[mask] == labels[mask]).sum())
      acc = correct/ total

    rows.append({
        "class_idx" : idx,
        "class_name" : class_name,
        "total" : int(total),
        "correct" : int(correct),
        "accuracy" : acc
    })
  return pd.DataFrame(rows).sort_values("accuracy", ascending=True)
```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np

epochs = 15
patience = 3
min_delta = 0.0005

best_val_acc = 0.0
best_epoch = 0
epochs_no_improve = 0

phase15_changed = False
phase2_changed = False

phase15_trigger_clean_train_acc = 0.86
phase2_trigger_clean_train_acc = 0.94

mixup_alpha = 0.2

center_loss_fn = None
center_optimizer = None
center_loss_weight = 0.0

save_path = "/content/drive/MyDrive/models/officehome_convnext_phase_center_best.pt"
os.makedirs(os.path.dirname(save_path), exist_ok=True)

history = {
    "epoch": [],
    "phase": [],
    "mixup_alpha": [],

    "train_loss_mixup": [],
    "train_acc_ref": [],
    "train_time": [],

    "clean_train_loss": [],
    "clean_train_acc": [],
    "clean_train_time": [],

    "val_loss": [],
    "val_acc": [],
    "val_top3": [],
    "val_time": [],

    "overfit_gap_clean": [],
    "lr": [],
}

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode="max",
    factor=0.5,
    patience=1
)

for epoch in range(epochs):
    print(f"\n===== Epoch [{epoch + 1}/{epochs}] =====")

    if phase2_changed:
        current_phase = 2
    elif phase15_changed:
        current_phase = 1.5
    else:
        current_phase = 1

    print("현재 Phase:", current_phase)
    print("현재 MixUp alpha:", mixup_alpha)

    # =========================================================
    # 1. Train
    # =========================================================
    if phase2_changed:
        train_loss, train_acc_ref, ce_loss, center_loss_value, train_time = train_one_epoch_centerloss_dualview(
            model=model,
            train_loader=train_loader,
            criterion=criterion,
            center_loss_fn=center_loss_fn,
            optimizer=optimizer,
            center_optimizer=center_optimizer,
            device=device,
            center_loss_weight=center_loss_weight
        )

        mix_loss = ce_loss
        cons_loss = center_loss_value

    else:
        train_loss, train_acc_ref, mix_loss, cons_loss, train_time = train_one_epoch_mixup_global_consistency(
            model=model,
            train_loader=train_loader,
            criterion=criterion,
            optimizer=optimizer,
            device=device,
            alpha=mixup_alpha,
            consistency_weight=0.1,
            temperature=1.0
        )

    # =========================================================
    # 2. Validation
    # =========================================================
    val_loss, val_acc, val_top3, val_preds, val_labels, val_time = evaluate(
        model=model,
        val_loader=val_loader,
        criterion=criterion,
        device=device
    )

    # =========================================================
    # 3. Clean train acc
    # =========================================================
    clean_train_loss, clean_train_acc, clean_train_time = evaluate_clean_train_acc(
        model=model,
        train_loader=clean_train_loader,
        criterion=criterion,
        device=device,
        max_batches=150
    )

    current_lr = optimizer.param_groups[0]["lr"]
    overfit_gap_clean = clean_train_acc - val_acc

    print(
        f"\nEpoch [{epoch + 1}/{epochs}] 결과\n"
        f"Phase: {current_phase} | MixUp alpha: {mixup_alpha}\n"
        f"Train Total Loss: {train_loss:.4f} | Aux1 Loss: {mix_loss:.4f} | Aux2 Loss: {cons_loss:.4f} | Acc Ref: {train_acc_ref:.4f}\n"
        f"Clean Train Loss: {clean_train_loss:.4f} | Clean Train Acc: {clean_train_acc:.4f}\n"
        f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f} | Val Top3: {val_top3:.4f}\n"
        f"Clean Overfit Gap: {overfit_gap_clean:.4f} | LR: {current_lr}"
    )

    # =========================================================
    # 4. History
    # =========================================================
    history["epoch"].append(epoch + 1)
    history["phase"].append(current_phase)
    history["mixup_alpha"].append(mixup_alpha)

    history["train_loss_mixup"].append(train_loss)
    history["train_acc_ref"].append(train_acc_ref)
    history["train_time"].append(train_time)

    history["clean_train_loss"].append(clean_train_loss)
    history["clean_train_acc"].append(clean_train_acc)
    history["clean_train_time"].append(clean_train_time)

    history["val_loss"].append(val_loss)
    history["val_acc"].append(val_acc)
    history["val_top3"].append(val_top3)
    history["val_time"].append(val_time)

    history["overfit_gap_clean"].append(overfit_gap_clean)
    history["lr"].append(current_lr)

    # =========================================================
    # 5. W&B log
    # =========================================================
    try:
        import wandb

        if wandb.run is not None:
            wandb.log({
                "epoch": epoch + 1,
                "phase": current_phase,

                "mixup/alpha": mixup_alpha,

                "train/total_loss": train_loss,
                "train/acc_ref": train_acc_ref,
                "train/time_sec": train_time,

                "train/aux_loss_1": mix_loss,
                "train/aux_loss_2": cons_loss,
                "train/center_loss_weight": center_loss_weight,

                "clean_train/loss": clean_train_loss,
                "clean_train/acc": clean_train_acc,
                "clean_train/time_sec": clean_train_time,

                "val/loss": val_loss,
                "val/acc_top1": val_acc,
                "val/acc_top3": val_top3,
                "val/time_sec": val_time,

                "diagnostics/overfit_gap_clean": overfit_gap_clean,
                "optimizer/lr": current_lr,
            })
    except Exception as e:
        print("W&B log 실패:", e)

    # =========================================================
    # 6. Best model save
    # =========================================================
    if val_acc > best_val_acc + min_delta:
        best_val_acc = val_acc
        best_epoch = epoch + 1
        epochs_no_improve = 0

        save_dict = {
            "model_state_dict": model.state_dict(),
            "classes": classes,
            "class_to_idx": class_to_idx,
            "val_acc": best_val_acc,
            "val_loss": val_loss,
            "val_top3": val_top3,
            "clean_train_acc": clean_train_acc,
            "epoch": best_epoch,
            "num_classes": num_classes,
            "model_name": "convnext_tiny",
            "phase": current_phase,
            "mixup_alpha": mixup_alpha,
            "phase15_changed": phase15_changed,
            "phase2_changed": phase2_changed,
            "center_loss_weight": center_loss_weight,
            "center_loss_used": phase2_changed,
            "input_size": "224x384",
            "note": "ConvNeXt + Phase + Hard Phase 2 + Center Loss."
        }

        if center_loss_fn is not None:
            save_dict["center_loss_state_dict"] = center_loss_fn.state_dict()

        torch.save(save_dict, save_path)

        print(f"Best model 저장 완료 | Epoch: {best_epoch} | Val Acc: {best_val_acc:.4f}")

    else:
        epochs_no_improve += 1
        print(f"성능 개선 없음: {epochs_no_improve}/{patience}")

    scheduler.step(val_acc)

    # =========================================================
    # 7. Phase 1.5
    # =========================================================
    if (clean_train_acc >= phase15_trigger_clean_train_acc) and (not phase15_changed):
        print("\n" + "=" * 60)
        print(f"Phase 1.5 전환 조건 충족: clean_train_acc={clean_train_acc:.4f} >= {phase15_trigger_clean_train_acc}")
        print("외우기 전 조기개입을 시작합니다.")
        print("하단부는 가리지 않습니다. TV/Monitor/Computer 구조 정보는 보존합니다.")
        print("=" * 60)

        train_dataset.transform_global = phase15_train_transform
        train_dataset.transform_local = consistency_local_transform
        print("train_dataset global/local transform 변경 완료")

        model = add_dropout_to_convnext_classifier(
            model,
            dropout_p=0.1
        )
        model = model.to(device)

        criterion = nn.CrossEntropyLoss(label_smoothing=0.08)
        print("criterion 변경: label_smoothing=0.08")

        for param in model.parameters():
            param.requires_grad = True

        optimizer = optim.AdamW(
            model.parameters(),
            lr=7e-5,
            weight_decay=1e-4
        )
        print("optimizer 변경: 전체 모델 학습 유지, lr=7e-5, weight_decay=1e-4")

        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
            optimizer,
            mode="max",
            factor=0.5,
            patience=1
        )
        print("scheduler 재설정 완료")

        mixup_alpha = 0.3
        print("MixUp alpha 변경: 0.3")

        phase15_changed = True
        epochs_no_improve = 0
        print("Phase 1.5 전환 완료, epochs_no_improve 초기화")

    # =========================================================
    # 8. Hard Phase 2 + Center Loss
    # =========================================================
    if (clean_train_acc >= phase2_trigger_clean_train_acc) and phase15_changed and (not phase2_changed):
        print("\n" + "=" * 60)
        print(f"Hard Phase 2 전환 조건 충족: clean_train_acc={clean_train_acc:.4f} >= {phase2_trigger_clean_train_acc}")
        print("저적중률 클래스를 더 자주 보고 Center Loss로 feature를 모읍니다.")
        print("=" * 60)

        priority_names = [
            "Computer",
            "Monitor",
            "TV",
            "Laptop",
            "Printer",
            "Keyboard"
        ]

        hard_class_names, current_class_acc_df = get_hard_classes_from_val(
            val_labels=val_labels,
            val_preds=val_preds,
            classes=classes,
            acc_threshold=0.82,
            max_classes=8,
            priority_names=priority_names
        )

        print("Hard classes:")
        print(hard_class_names)

        display(
            current_class_acc_df[
                current_class_acc_df["class_name"].isin(hard_class_names)
            ].sort_values("accuracy")
        )

        train_dataset.transform_global = phase15_train_transform
        train_dataset.transform_local = consistency_local_transform
        print("train_dataset transform 변경 완료")

        hard_sampler = make_hard_class_sampler(
            dataset=train_dataset,
            hard_class_names=hard_class_names,
            class_to_idx=class_to_idx,
            hard_sample_weight=4.0,
            normal_sample_weight=1.0
        )

        train_loader = DataLoader(
            train_dataset,
            batch_size=batch_size,
            sampler=hard_sampler,
            shuffle=False,
            num_workers=0,
            pin_memory=True if torch.cuda.is_available() else False
        )

        print("Hard class weighted train_loader 생성 완료")

        hard_class_weights = make_hard_class_weight_tensor(
            num_classes=num_classes,
            classes=classes,
            hard_class_names=hard_class_names,
            device=device,
            hard_loss_weight=1.4,
            normal_loss_weight=1.0
        )

        criterion = nn.CrossEntropyLoss(
            weight=hard_class_weights,
            label_smoothing=0.05
        )

        print("Hard class weighted criterion 생성 완료")

        model = freeze_convnext_front_keep_last(model)
        model = model.to(device)

        feat_dim = get_convnext_feature_dim(model)

        center_loss_fn = CenterLoss(
            num_classes=num_classes,
            feat_dim=feat_dim,
            device=device
        )

        center_optimizer = optim.SGD(
            center_loss_fn.parameters(),
            lr=0.5
        )

        center_loss_weight = 0.02

        print("Center Loss 준비 완료")
        print("feat_dim:", feat_dim)
        print("center_loss_weight:", center_loss_weight)

        trainable_params = [
            p for p in model.parameters()
            if p.requires_grad
        ]

        optimizer = optim.AdamW(
            trainable_params,
            lr=3e-5,
            weight_decay=1e-4
        )

        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
            optimizer,
            mode="max",
            factor=0.5,
            patience=1
        )

        mixup_alpha = 0.0

        phase2_changed = True
        epochs_no_improve = 0

        print("Hard Phase 2 + Center Loss 전환 완료")
        print("MixUp alpha:", mixup_alpha)

    # =========================================================
    # 9. Early stopping
    # =========================================================
    if epochs_no_improve >= patience:
        print(f"Early stopping 발생 | Epoch {epoch + 1}")
        break


# =========================================================
# Final summary
# =========================================================
class_acc_df = make_class_accuracy_table(
    labels=val_labels,
    preds=val_preds,
    classes=classes
)

try:
    import wandb

    if wandb.run is not None:
        wandb.run.summary["best_val_acc"] = best_val_acc
        wandb.run.summary["best_epoch"] = best_epoch
        wandb.run.summary["final_val_acc"] = val_acc
        wandb.run.summary["final_clean_train_acc"] = clean_train_acc

        wandb.log({
            "tables/class_accuracy": wandb.Table(dataframe=class_acc_df)
        })

        wandb.finish()

except Exception as e:
    print("W&B finish 실패:", e)

print("\n학습 완료")
print("Best Val Acc:", round(best_val_acc, 4))
print("Best Epoch:", best_epoch)
print("저장 경로:", save_path)
print("Phase 1.5 전환 여부:", phase15_changed)
print("Phase 2 전환 여부:", phase2_changed)
```
